```python
# Initialize sentiment analysis pipeline
sentiment_analysis = pipeline("sentiment-analysis", model="siebert/sentiment-roberta-large-english")

# List of app IDs to scrape
app_ids =[
    'com.twgbd.dims',
    'com.medexplus.digitalitsheba.app',
    'com.bizit.doctor.appointment',
    'com.simpleappsvault.temperaturethermometerdiary',
    'com.msd.professional',
    'com.bondhu.daktar.daktarbondhu',
    'com.arbondhu',
    'com.media365ltd.doctime',
    'com.msd.consumer',
    'com.makio.medica',
    'com.nhcps.medicode',
    'com.healthtipsbd.medicineapp',
    'da.bd.drug.directory',
    'com.classnotebd.docterapp',
    'com.blood.sugar.tracker.glucometer.Diary.test.diabetes.checker.info.glucose.convert.logger.health.fitness.monitor.history',
    'com.classnotesbd.kobiraji_chikitsa_book',
    'xray.body.scanne.ronasoftstudiosr.simulator.filter.camera',
    'com.ronasoftstudios.earmaxfxpro',
    'com.osudpotro',
    'com.codersbucket.surokkha_app',
    'com.itmedicus.dimsvet',
    'in.naskar.achal.homoeopathy',
    'com.bissoy.callomy3DViewer3',
    'com.drugscom.app',
    'com.AnatomyLearning.Anatomy3DViewer3',
    'org.narayanahealth.nhcare',
    'com.Medical.Know.Pregnant.Mother',
    'net.medshr.android',
    'com.classnotebd.imagetotextword',
    'com.classnotesbd.homeopathy_guide_book',
    'com.mobisystems.msdict.embedded.wireless.oxford.concisemedical',
    'com.healtho_max.app.production',
    'com.catfishanimationstudio.MuscularSystemLtreatment',
    'sqh.jotno.patient',
    'com.rtrg.xraywow',
    'com.sevenonelab.harbal_medicine_treatment',
    'com.catfishanimationstudio.SkeletalSystemPreview',
    'me.sinacare.app.android',
    'com.pulsetechCalendar',
    'com.ltd.medicart',
    'com.sevenonelab.bangla_medicine_info',
    'com.snonapps.aidoctor',
    'com.lbrc.PeriodCalendar',
    'com.EAGINsoftware.dejaloYa',
    'io.healthos.androidecom',
    'com.digitalbithub.shen',
    'com.farmer.disease.animal.farm'
]    # Add more app IDs as needed

# Create an empty DataFrame to store all reviews
all_reviews = pd.DataFrame()

# Loop through each app ID and scrape reviews
for app_id in app_ids:
    print(f"Scraping reviews for {app_id}...")
```

```python
# Scraping reviews for the current app
hk_project = reviews_all(app_id, sleep_milliseconds=0, lang='en', country='US', sort=Sort.NEWEST)
```

```python
# Normalize the JSON data to a DataFrame
df = pd.json_normalize(hk_project)
```

```python
# Perform sentiment analysis
df['content'] = df['content'].astype('str')
```

```python
df['result'] = df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
# Extract sentiment and score
df['sentiment'] = df['result'].apply(lambda x: (x[0]['label']))
```

```python
df['score'] = df['result'].apply(lambda x: (x[0]['score']))
```

```python
 # Add a column for the app ID
 df['app_id'] = app_id
```

```python
 # Append the current app's reviews to the all_reviews DataFrame
all_reviews = pd.concat([all_reviews, df], ignore_index=True)
```

```python
# Analyze overall sentiment
print("Average sentiment score:", all_reviews['score'].mean())
```

```python
print("Sentiment value counts:\n", all_reviews['sentiment'].value_counts(normalize=True))
```

```python
# Create histogram for sentiment distribution
fig = px.histogram(all_reviews, x='sentiment', color='sentiment', text_auto=True)
fig.show()
```

```python
# Save all reviews to a CSV file
all_reviews.to_csv('sentiment2.csv', index=False)
```
