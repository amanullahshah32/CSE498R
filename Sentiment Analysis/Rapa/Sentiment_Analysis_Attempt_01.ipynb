```python
import pandas as pd
import re
```

```python
def is_mixed_language(text):
    # Regex to find Bengali characters
    bengali_pattern = re.compile('[\u0980-\u09FF]')
    # Check if both Bengali and English characters are present
    return bool(bengali_pattern.search(text)) and bool(re.search('[a-zA-Z]', text))

# Apply the function to the review_description column
mixed_language_reviews = df[df['review_description'].apply(is_mixed_language)]

# Print the number of rows containing mixed language reviews
print(f'Number of mixed language reviews: {mixed_language_reviews.shape[0]}')
```

```python
# Remove mixed language reviews from the DataFrame
df_cleaned = df[~df['review_description'].apply(is_mixed_language)]

# Optionally, you can save the cleaned DataFrame to a new CSV file
df_cleaned.to_csv('cleaned_reviews.csv', index=False)  # Save the cleaned data to 'cleaned_reviews.csv'

# Print the number of rows in the cleaned DataFrame
print(f'Number of rows after removing mixed language reviews: {df_cleaned.shape[0]}')
```

```python
import pandas as pd
import re


# Function to check for emojis
def contains_emojis(text):
    # Regex pattern for emojis
    emoji_pattern = re.compile("[\U0001F600-\U0001F64F"  # emoticons
                                 "\U0001F300-\U0001F5FF"  # symbols & pictographs
                                 "\U0001F680-\U0001F6FF"  # transport & map symbols
                                 "\U0001F700-\U0001F77F"  # alchemical symbols
                                 "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                                 "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                                 "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                                 "\U0001FA00-\U0001FA6F"  # Chess Symbols
                                 "\U00002702-\U000027B0"  # Miscellaneous Symbols
                                 "\U000024C2-\U0001F251"  # Enclosed Characters
                                 "]+", flags=re.UNICODE)
    return bool(emoji_pattern.search(text))

# Remove rows with emojis from the review_description column
df_cleaned_no_emojis = df[~df['review_description'].apply(contains_emojis)]

# Optionally, save the cleaned DataFrame to a new CSV file
df_cleaned_no_emojis.to_csv('cleaned_reviews_no_emojis.csv', index=False)  # Save to 'cleaned_reviews_no_emojis.csv'

# Print the number of rows in the cleaned DataFrame without emojis
print(f'Number of rows after removing reviews with emojis: {df_cleaned_no_emojis.shape[0]}')
```

## Clean the dataset and keep only reviews and ratings

```python
# Keep only 'review_description' and 'rating' columns
df = df[['review_description', 'rating']]

# Save the modified dataset if needed
df.to_csv('modified_dataset.csv', index=False)  # or to_excel, to_json, etc., depending on your preferred output format

# Display the modified DataFrame
print(df.head(), df.shape)
```

## **Import Libraries**

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
import torch
from torch.utils.data import DataLoader, TensorDataset
```

## **Initialize Tokenizer and Model**

```python
# Load the pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Load the pre-trained BERT model for sequence classification (sentiment analysis)
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)  # 3 labels: positive, neutral, negative
```

## **Tokenize the Text**

```python
# Define a maximum token length
max_length = 128

# Tokenize the review text
tokens = tokenizer(
    df['review_description'].tolist(),
    max_length=max_length,
    padding=True,
    truncation=True,
    return_tensors='pt'
)
```

## **Create DataLoader**

```python
# Move tokens to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokens['input_ids'] = tokens['input_ids'].to(device)
tokens['attention_mask'] = tokens['attention_mask'].to(device)

# Create a TensorDataset and DataLoader for batch processing
dataset = TensorDataset(tokens['input_ids'], tokens['attention_mask'])
dataloader = DataLoader(dataset, batch_size=16)  # Adjust batch size as needed
```

## Run Sentiment Analysis

```python
# Step 7: Run Sentiment Analysis on GPU
model.to(device)
model.eval()

# Initialize an empty list to store predicted sentiment labels
sentiment_labels = []

# Run inference on each batch
with torch.no_grad():
    for batch in dataloader:
        input_ids, attention_mask = batch

        # Move data to GPU
        input_ids = input_ids.to(device)
        attention_mask = attention_mask.to(device)

        # Get model outputs
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        # Get predicted label (sentiment) for each review
        predictions = torch.argmax(logits, dim=1)
        sentiment_labels.extend(predictions.cpu())  # Move to CPU for easy manipulation

# Confirm that sentiment_labels has values
print(f"Number of sentiment labels: {len(sentiment_labels)}")
```

## **Interpret Results**

```python
# Map numeric labels to text labels
sentiment_map = {0: "negative", 1: "neutral", 2: "positive"}
df['sentiment'] = [sentiment_map[label.item()] for label in sentiment_labels]

# Display the first few rows with sentiment results
df[['review_description', 'sentiment']].head()
```
